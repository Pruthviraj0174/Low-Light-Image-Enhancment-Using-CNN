```python
import os
import tensorflow as tf
import cv2
import numpy as np

target_size = (256, 256) 

image_extensions = ['.png', '.jpg', '.jpeg']

def is_image_file(filename):

    return any(filename.lower().endswith(ext) for ext in image_extensions)


def load_and_preprocess_image(image_path):

    image = cv2.imread(image_path)
    
  
    if image is None:
        print(f"Failed to load {image_path}")  # Debug message if image loading fails
        return None

    image = cv2.resize(image, target_size)

    image = image / 255.0
    
    # Convert the image to a TensorFlow tensor
    image = tf.convert_to_tensor(image, dtype=tf.float32)
    
    return image

image_data = {}

folders = [
    "training/high", "training/low",
    "testing/high", "testing/low"
]

for folder in folders:
    print(f"Loading images from {folder}...")
    folder_images = []  
    
    for file_name in os.listdir(folder):
        file_path = os.path.join(folder, file_name)  
 
        if is_image_file(file_name):

            preprocessed_image = load_and_preprocess_image(file_path)
            if preprocessed_image is not None:
                folder_images.append(preprocessed_image)  
    

    image_data[folder] = folder_images
    print(f"Loaded {len(folder_images)} preprocessed images from {folder}")
```

```python
import tensorflow as tf

def create_dataset(image_dict):
    image_tensors = []
    labels = []
    for label, images in image_dict.items():
        for image in images:
            image_tensors.append(image)
            labels.append(label)  

    dataset = tf.data.Dataset.from_tensor_slices((image_tensors, labels))
    return dataset


dataset = create_dataset(image_data)
```

```python
import os
import tensorflow as tf
from tensorflow.keras.preprocessing.image import load_img, img_to_array

low_light_dir = 'training/low'
high_light_dir = 'training/high'

def load_and_preprocess_image(image_path, target_size=(256, 256)):
    
    img = load_img(image_path, target_size=target_size)
    img_array = img_to_array(img) / 255.0  
    return img_array

def create_paired_dataset(low_dir, high_dir, target_size=(256, 256)):
    paired_images = []

   
    low_images = sorted(os.listdir(low_dir))
    high_images = sorted(os.listdir(high_dir))


    assert len(low_images) == len(high_images), "Mismatch in number of images."

   
    for low_img_name, high_img_name in zip(low_images, high_images):
        low_img_path = os.path.join(low_dir, low_img_name)
        high_img_path = os.path.join(high_dir, high_img_name)

        low_img = load_and_preprocess_image(low_img_path, target_size)
        high_img = load_and_preprocess_image(high_img_path, target_size)

     
        paired_images.append((low_img, high_img))

    low_images, high_images = zip(*paired_images)
    dataset = tf.data.Dataset.from_tensor_slices((list(low_images), list(high_images)))

    return dataset


paired_dataset = create_paired_dataset(low_light_dir, high_light_dir)


paired_dataset = paired_dataset.shuffle(buffer_size=100).batch(32)
```

```python
from tensorflow.keras import layers, models

def create_cnn_model(input_shape=(256, 256, 3)):
    model = models.Sequential([
        layers.InputLayer(input_shape=input_shape),
        layers.Conv2D(32, (3, 3), activation='relu', padding='same'),
        layers.MaxPooling2D((2, 2)),
        layers.Conv2D(64, (3, 3), activation='relu', padding='same'),
        layers.MaxPooling2D((2, 2)),
        layers.Conv2D(128, (3, 3), activation='relu', padding='same'),
        layers.UpSampling2D((2, 2)),
        layers.Conv2D(64, (3, 3), activation='relu', padding='same'),
        layers.UpSampling2D((2, 2)),
        layers.Conv2D(3, (3, 3), activation='sigmoid', padding='same')
    ])

    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
cnn_model = create_cnn_model()
cnn_model.fit(paired_dataset, epochs=20)
```

```python

test_dataset = create_paired_dataset('testing/low', 'testing/high')
test_dataset = test_dataset.batch(32)
test_loss = cnn_model.evaluate(test_dataset)
print(f'Test Loss: {test_loss}')
```

```python
import matplotlib.pyplot as plt
for low_light_images, high_light_images in test_dataset.take(1):
    predicted_images = cnn_model.predict(low_light_images)
    for i in range(3):  
        plt.figure(figsize=(12, 4))

        plt.subplot(1, 3, 1)
        plt.title("Low-Light Image")
        plt.imshow(low_light_images[i].numpy())
        plt.axis('off')

        # Predicted (Enhanced) image
        plt.subplot(1, 3, 2)
        plt.title("Enhanced Image")
        plt.imshow(predicted_images[i])
        plt.axis('off')

        plt.subplot(1, 3, 3)
        plt.title("Ground Truth (High-Light)")
        plt.imshow(high_light_images[i].numpy())
        plt.axis('off')

        plt.show()
```
