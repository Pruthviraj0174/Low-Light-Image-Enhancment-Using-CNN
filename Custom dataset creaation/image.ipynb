```python
from imutils import paths
import cv2
import numpy as np
import random
```

```python
imagePaths= list(paths.list_images("images"))
imagePaths
```

```python
def add_gaussian_noise(image):
    image = image.astype(np.float32)
    shape = image.shape[:2]

    mean = 0
    var = random.uniform(0,0.1)
    sigma = var ** 0.5
    gamma = 0.25
    alpha = 0.75
    beta = 1 - alpha

    gaussian = np.random.normal(loc=mean, scale = sigma, size = (shape[0], shape[1], 1)).astype(np.float32)
    gaussian = np.concatenate((gaussian, gaussian, gaussian), axis = 2)
    #gaussian_img = image * 0.75 + 0.25 * gaussian + 0.25
    gaussian_img = cv2.addWeighted(image, alpha, beta * gaussian, beta, gamma)

    return gaussian_img
 

i=0

for imagePath in imagePaths:
    img = cv2.imread(imagePath)
    

    # Create a black image with the same dimensions as the original image
    black_img = np.zeros(img.shape, dtype=np.uint8)

    # Blend the original image with the black image
    blended_img = cv2.addWeighted(img, 0.3, black_img, 0.9, 0)

    cv2.imwrite(f'dark_images\{i}'+' dark_image.jpg', add_gaussian_noise(blended_img))
    i=i+1
```
